```python
import numpy as np
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
size = 1000 * rng.randn(100)

plt.scatter(x, y, c=colors, s=size, alpha=0.3, cmap='viridis')

plt.colorbar()
```
