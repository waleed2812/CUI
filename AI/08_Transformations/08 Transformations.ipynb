```python
import numpy as np
from matplotlib import pyplot as plt
```

### Normal 30x30 grid. Ranging from -15 to 15

```python
grid = np.column_stack([[x, y] for x in range(-15, 15) for y in range(-15, 15)])
print(grid)
grid.shape
```

### Function to plot a grid

```python
def plot_grid(grid, title='900 Vectors Normal Representation in Space'):
    
    plt.figure(figsize=(8, 6), dpi=80)
    
    plt.title(title)

    plt.scatter(grid[0], grid[1])
```

### Normal Plotting of 30x30 grid. Each point generated by cartesian product of x and y

```python
plot_grid(grid)
```

### Matrix for Shear Transformation (Shear Matrix)

```python
shear_matrix = np.array([1, 1, 0, 1]).reshape(2, 2)
shear_matrix
```

#### Getting Shear Transformed Points

```python
sheared_grid = shear_matrix.dot(grid)
sheared_grid
```

#### Plotting

```python
plot_grid(sheared_grid, title='900 Vectors Shear Representation in Space')
```

# 3D Grid

```python
grid_3d = np.column_stack( [ [x, y, z] for x in range(-15, 15) for y in range(-15, 15) for z in range(-15, 15)])

print(grid_3d)
grid_3d.shape
```

```python
def plot_3D(grid_3d):

    fig = plt.figure(figsize=(10, 8), dpi=80)

    ax = plt.axes(projection='3d')

    ax.scatter3D(grid_3d[0], grid_3d[1], grid_3d[2], c=grid_3d[2], cmap='Greens');

plot_3D(grid_3d)
```

```python
shear_matrix_3d = np.array([1, 1, 0, 1, 0, 1, 1, 0, 1]).reshape(3, 3)
shear_matrix_3d
```

```python
sheared_grid_3d = shear_matrix_3d.dot(grid_3d)
sheared_grid_3d
```

```python
plot_3D(sheared_grid_3d)
```
